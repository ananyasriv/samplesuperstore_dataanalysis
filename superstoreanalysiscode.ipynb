```python
# Scatter plot of Sales vs Profit
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales'], df['Profit'], color='blue', alpha=0.5)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit')
plt.show()
```

```python
# Bar plot of Sales for each Category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sales', data=df, ci=None)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()

# Bar plot of Profit for each Category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Profit', data=df, ci=None)
plt.xlabel('Category')
plt.ylabel('Profit')
plt.title('Profit by Category')
plt.show()
```

```python
# Box plot of Profit for each Region
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Profit', data=df)
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit by Region')
plt.show()
```

```python
# Pairplot for overall analysis of numerical columns
sns.pairplot(df, diag_kind='kde')
plt.show()
```
